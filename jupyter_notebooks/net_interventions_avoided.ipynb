```python
import dcurves
```

```python
df_binary = dcurves.load_test_data.load_binary_df()
```

```python
dca_binary_df = dcurves.dca(
    data = df_binary,
    outcome = 'cancer',
    predictors = ['famhistory']
           )
```

```python

all_records = dca_binary_df[dca_binary_df['variable'] == 'all']
all_records = all_records[['threshold', 'net_benefit']]
```

```python
all_records = all_records.rename(columns = {'net_benefit':'net_benefit_all'})
```

```python
dca_df_allmerged = dca_binary_df.merge(all_records, on='threshold')
```

```python
nper = 100
```

```python
dca_df_allmerged['net_intervention_avoided'] = (dca_df_allmerged['net_benefit'] \
                                                - dca_df_allmerged['net_benefit_all']) \
                                                / (dca_df_allmerged['threshold'] \
                                                / (1 - dca_df_allmerged['threshold']))*nper
```
